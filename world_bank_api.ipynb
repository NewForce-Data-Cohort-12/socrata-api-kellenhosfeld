<h2>Use the API to get all available data for the GDP per capita, PPP (constant 2021 international $) indicator.</h2>

Hint: this indicator has code "NY.GDP.PCAP.PP.KD". Adjust the query parameters so that you can retrieve all available rows. Convert the results to a DataFrame.

```python
import requests
import pandas as pd
import time
```

```python
url = "https://api.worldbank.org/v2/country/all/indicator/NY.GDP.PCAP.PP.KD?per_page=10000&format=json"
url2 = "https://api.worldbank.org/v2/country/all/indicator/NY.GDP.PCAP.PP.KD?per_page=10000&page=2&format=json"

# Send GET request
response = requests.get(url)
response2 = requests.get(url2)

# Check response status
if response.status_code == 200:
    print("Request successful!")
    #print(response2.json())
    gdp_p1 = response.json()
    gdp_p2 = response2.json()
    gdp_p1_json = pd.json_normalize(gdp_p1[1])
    gdp_p2_json = pd.json_normalize(gdp_p2[1])
    gdp_df = pd.concat([gdp_p1_json, gdp_p2_json])
else:
    print("Request failed with status code:", response.status_code)
```

```python
gdp_df.head()
```

<h2>Now, use the API to get all available data for Life expectancy at birth, total (years). This indicator has code "SP.DYN.LE00.IN". Again, convert the results to a DataFrame.</h2>

```python
url = "https://api.worldbank.org/v2/country/all/indicator/SP.DYN.LE00.IN?per_page=10000&format=json"
url2 = "https://api.worldbank.org/v2/country/all/indicator/SP.DYN.LE00.IN?per_page=10000&page=2&format=json"

# Send GET request
response = requests.get(url)
response2 = requests.get(url2)

# Check response status
if response.status_code == 200:
    print("Request successful!")
    #print(response2.json())
    lab_p1 = response.json()
    lab_p2 = response2.json()
    lab_p1_json = pd.json_normalize(lab_p1[1])
    lab_p2_json = pd.json_normalize(lab_p2[1])
    lab_df = pd.concat([lab_p1_json, lab_p2_json])
else:
    print("Request failed with status code:", response.status_code)
```

```python
lab_df.tail()
```

<h2>Merge the two results DataFrames together. You may want to rename or drop columns prior to merging.</h2>

```python
gdp_df = gdp_df.drop(columns=['unit', 'obs_status', 'decimal', 'indicator.id', 'indicator.value'])
```

```python
gdp_df = gdp_df[['date', 'country.value', 'country.id', 'countryiso3code', 'value']]
gdp_df.head()
```

```python
gdp_df.columns = ['year', 'country', 'country_id', 'country_iso3', 'gdp_per_capita']
```

```python
gdp_df.head()
```

```python
lab_df = lab_df.drop(columns=['countryiso3code', 'unit', 'obs_status', 'decimal', 'indicator.id', 'indicator.value', 'country.value'])
```

```python
lab_df.columns = ['year', 'life_expectancy_at_birth', 'country_id']
lab_df.head()
```

```python
gdp_lab_df = pd.merge(gdp_df, lab_df, on = ['year', 'country_id'])
gdp_lab_df.head()
```

<h2>You can also get more information about the available countries (region, capital city, income level classification, etc.) by using the Country API.</h2>

Use this API to pull in all available data. Merge this with your other datasets. Use this to now remove the rows that correspond to regions and not countries.

```python
url = "https://api.worldbank.org/v2/country?per_page=300&format=json"

# Send GET request
response = requests.get(url)

# Check response status
if response.status_code == 200:
    print("Request successful!")
    print(response.json())
    countries = response.json()
    countries_df = pd.json_normalize(countries[1])
else:
    print("Request failed with status code:", response.status_code)
```

```python
countries_df.head()
```

```python
gdp_lab_countries = gdp_lab_df.merge(countries_df, left_on='country_id', right_on='iso2Code')
gdp_lab_countries.head()
```

```python
print(gdp_lab_countries['region.value'].unique())
```

```python
print(gdp_lab_countries.loc[gdp_lab_countries['region.value'] == 'Aggregates']['country'].unique())
```

```python
gdp_lab_countries = gdp_lab_countries.loc[gdp_lab_countries['region.value'] != 'Aggregates']
print(gdp_lab_countries['country'].unique())
```

```python
gdp_lab_countries = gdp_lab_countries.reset_index(drop=True)
```
